```python
import pandas as pd
import numpy as np

X_train = pd.read_csv('./datasets/TON-IoT/ton_iot_train.csv')
X_test = pd.read_csv('./datasets/TON-IoT/ton_iot_test.csv')
X = pd.concat([X_train, X_test])
X.sort_values(by='ID', ascending=True, inplace=True)
```

```python
X
```

```python
type_map = {'normal': 0, 'backdoor': 1, 'ddos': 2, 'dos': 3, 'injection': 4, 'password': 5, 'ransomware': 6, 'scanning': 7, 'xss': 8, 'mitm': 9}
```

```python
nodes = np.unique(np.concatenate([X['src_ip'].values, X['dst_ip'].values]))
```

```python
nodes_num = len(nodes)
links_num = len(X)
```

```python
type_num = []
for i in range(len(type_map.keys())):
    type_num.append({'type': list(type_map.keys())[i],'num': len(X[X['type'] == i])})
```

```python
type_num
```

```python
dataset_description = {
    'nodes_num': nodes_num,
    'links_num': links_num,
    'type_num': type_num
}
```

```python
import json
json.dump(dataset_description, open('./datasets/TON-IoT/dataset_description.json', 'w'))
```
