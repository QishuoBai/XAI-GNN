```python
import pandas as pd
import numpy as np

X_train = pd.read_csv('./datasets/TON-IoT/ton_iot_train.csv')
X_test = pd.read_csv('./datasets/TON-IoT/ton_iot_test.csv')
X = pd.concat([X_train, X_test])
X.sort_values(by='ID', ascending=True, inplace=True)
```

```python
X
```

```python
explain_id = 0
```

```python
# 找到id对应两跳的数据
expand_round = 2
ids = np.array([explain_id])
for _ in range(expand_round):
    cur_nodes = np.unique(np.concatenate([X[X['ID'].isin(ids)]['src_ip'].values, X[X['ID'].isin(ids)]['dst_ip'].values]))
    not_in_ids = X[~X['ID'].isin(ids)]
    not_in_ids = not_in_ids[(not_in_ids['src_ip'].isin(cur_nodes)) | (not_in_ids['dst_ip'].isin(cur_nodes))]
    if not_in_ids.shape[0] == 0:
        break
    ids = np.concatenate([ids, not_in_ids['ID'].values])
```

```python
ids
```

```python
sub_X = X[X['ID'].isin(ids)].copy()
```

```python
sub_X
```

```python
norm_cols = X.columns.to_list()[3:-2]
X['h'] = X[norm_cols].values.tolist()
sub_X['h'] = sub_X[norm_cols].values.tolist()
```

```python
norm_cols
```

```python
from models.GraphSAGE import SAGE
from models.MLP import MLPPredictor
import torch.nn.functional as F
import torch.nn as nn
import networkx as nx
from dgl import from_networkx
import torch as th

eattrs = ['ID', 'h', 'label', 'type']
G = nx.from_pandas_edgelist(sub_X, 'src_ip', 'dst_ip', eattrs, create_using=nx.MultiGraph)
G = G.to_directed()
G = from_networkx(G, edge_attrs=eattrs)
G.ndata['h'] = th.ones(G.num_nodes(), G.edata['h'].shape[1])

feature_mask = nn.Parameter((th.ones(G.edata['h'].shape[1]) * 0.5).unsqueeze(0), requires_grad=True)
edge_mask = nn.Parameter((th.ones(G.edata['h'].shape[0]) * 0.5).unsqueeze(1), requires_grad=True)

class Model(nn.Module):
    def __init__(self, ndim_in, ndim_out, edim, activation, dropout, bin=False):
        super(Model, self).__init__()
        self.gnn = SAGE(ndim_in, ndim_out, edim, activation, dropout)
        if bin:
            self.pred = MLPPredictor(ndim_out, edim, 2)
        else:
            self.pred = MLPPredictor(ndim_out, edim, 10)
    def forward(self, g, nfeats, efeats):
        h = self.gnn(g, nfeats, efeats)
        return self.pred(g, h)
    
device = th.device('cuda' if th.cuda.is_available() else 'cpu')
print('Using device:', device)

params={
    'ndim_out': 128,
    'dropout': 0.2,
    'epochs': 500, # 2000
}
bin = False

# 加载模型
print('Loading model...')
model = Model(G.ndata['h'].shape[1], params['ndim_out'], G.edata['h'].shape[1], F.relu, params['dropout'], bin)
model.load_state_dict(th.load('pts/TON-IoT-p-ndim_out_128dropout_0.2epochs_500_multi(0.8653).pt'))

feature_mask = feature_mask.to(device)
edge_mask = edge_mask.to(device)
model = model.to(device)
G = G.to(device)
model.eval()

print(G.edata['h'])

pred_origin = model(G, G.ndata['h'], G.edata['h'])
pred_feature_masked = model(G, G.ndata['h'], G.edata['h'] * feature_mask)
pred_edge_masked = model(G, G.ndata['h'], G.edata['h'] * edge_mask)
print(pred_origin.argmax(dim=1))
print(G.edata['type'])
```

```python
def feature_mask_loss(g, pred_cur, feature_mask, edge_mask, eid):
    coeffs = {
        "pred": 3.0,
        "feat_size": 1.0,
        "feat_mask_ent": 0.1,
        "edge_size": 80.0,
        "edge_ent": 1.0
    }

    # MI loss
    label = g.edata['label'][eid]
    pred_loss = -th.log(pred_cur[eid][label])
    pred_loss = coeffs['pred'] * pred_loss

    # size
    feature_mask_sig = feature_mask.sigmoid()
    feature_mask_size_loss = coeffs['feat_size'] * th.mean(feature_mask_sig)
    edge_mask_sig = edge_mask.sigmoid()
    edge_mask_size_loss = coeffs['edge_size'] * th.mean(edge_mask_sig)

    # entropy
    feat_mask_ent = -feature_mask_sig * th.log(feature_mask_sig) - (1 - feature_mask_sig) * th.log(1 - feature_mask_sig)
    feat_mask_ent_loss = coeffs['feat_mask_ent'] * th.mean(feat_mask_ent)
    edge_mask_ent = -edge_mask_sig * th.log(edge_mask_sig) - (1 - edge_mask_sig) * th.log(1 - edge_mask_sig)
    edge_mask_ent_loss = coeffs['edge_ent'] * th.mean(edge_mask_ent)

    loss = pred_loss + feature_mask_size_loss + feat_mask_ent_loss + edge_mask_size_loss + edge_mask_ent_loss

    return loss
```

```python
result_df = pd.DataFrame(columns=['ID', 'type'])

result_df['ID'] = th.tensor(G.edata['ID']).cpu().detach().numpy()
result_df['type'] = th.tensor(G.edata['type']).cpu().detach().numpy()
result_df['pred_origin'] = pred_origin.cpu().detach().numpy().tolist()
result_df['pred_feature_masked'] = pred_feature_masked.cpu().detach().numpy().tolist()
result_df['pred_edge_masked'] = pred_edge_masked.cpu().detach().numpy().tolist()
```
