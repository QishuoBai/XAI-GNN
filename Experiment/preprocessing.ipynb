```python
# 为feature description添加range信息
import json
import numpy as np
feature_description = json.load(open('./datasets/TON-IoT/feature_description.json', 'r'))
for feature in feature_description:
    if feature['is_num']:
        min_val = data[feature['key']].min()
        max_val = data[feature['key']].max()
        feature['range'] = [min_val, max_val]
    else:
        values = np.unique(data[feature['key']])
        values.sort()
        feature['range'] = values.tolist()
json.dump(feature_description, open('./datasets/TON-IoT/feature_description.json', 'w'))
```
